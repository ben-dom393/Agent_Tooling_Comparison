# Exploratory Data Analysis
## 1. EDA Plan
- Load data and examine structure, types, missing values
- Explore target distribution (`y`) and class imbalance
- Analyse relationships between predictors (e.g. `age`, `job`) and the target
- Identify data quality issues (e.g. 'unknown' values, `pdays`)
- Flag potential leakage variables (e.g. `duration`)


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('bank-additional-full.csv', sep=';')
df.head()
```

## 2. Data Structure & Missing Values
Checking the shape of the dataset and occurrences of 'unknown' values.

```python
print(f"Data Shape: {df.shape}")

print("\nMissing values (NaNs):")
print(df.isna().sum().sum())

print("\n'unknown' values per categorical column:")
unknowns = df.isin(['unknown']).sum()
print(unknowns[unknowns > 0].sort_values(ascending=False))
```

## 3. Class Imbalance (Target Distribution)


```python
y_dist = df['y'].value_counts(normalize=True)
print("Target 'y' distribution:\n", y_dist)

plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='y', palette='Set2')
plt.title('Target Variable Distribution (y)')
plt.show()
```

## 4. Anomalies & Leakage Risks
- `duration`: Highly correlates with `y` but is unknown before the call. Huge leakage risk.
- `pdays`: The value 999 means 'not previously contacted'. It acts as a null/category indicator.

```python
# Check pdays anomaly
pdays_999 = sum(df['pdays'] == 999)
print(f"pdays = 999 count: {pdays_999} out of {len(df)} ({(pdays_999/len(df))*100:.1f}%)")

# Look at duration by target
plt.figure(figsize=(8,5))
sns.boxplot(data=df, x='y', y='duration', palette='Set2')
plt.title('Duration by Target (Warning: Leakage Risk)')
plt.show()
```

## 5. Bivariate Analysis

```python
plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='job', hue='y', palette='Set2')
plt.xticks(rotation=45)
plt.title('Target Distribution by Job')
plt.show()

plt.figure(figsize=(10, 5))
sns.boxplot(data=df, x='y', y='age', palette='Set2')
plt.title('Age Distribution by Target')
plt.show()
```
